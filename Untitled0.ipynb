```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

print("Dataset Loaded Successfully")
```

```python
from sklearn.model_selection import train_test_split

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print("Data Split Done")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier()

model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
```

```python
from sklearn.ensemble import BaggingClassifier

bag = BaggingClassifier()

bag.fit(X_train, y_train)

pred2 = bag.predict(X_test)

from sklearn.metrics import accuracy_score
print("Bagging Accuracy:", accuracy_score(y_test, pred2))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)

rf.fit(X_train, y_train)

pred3 = rf.predict(X_test)

from sklearn.metrics import accuracy_score
print("Random Forest Accuracy:", accuracy_score(y_test, pred3))
```

```python
from sklearn.ensemble import AdaBoostClassifier

adb = AdaBoostClassifier()

adb.fit(X_train, y_train)

pred4 = adb.predict(X_test)

from sklearn.metrics import accuracy_score
print("AdaBoost Accuracy:", accuracy_score(y_test, pred4))
```
